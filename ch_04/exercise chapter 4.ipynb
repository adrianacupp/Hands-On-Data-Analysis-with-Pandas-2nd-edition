3. Using the faang.csv file, group by the ticker and resample to monthly frequency.
Make the following aggregations:
- a) Mean of the opening price
- b) Maximum of the high price
- c) Minimum of the low price
- d) Mean of the closing price
- e) Sum of the volume traded

```python
faang = pd.read_csv('exercises/faang.csv', index_col='date', parse_dates=True)
faang.head()
```

```python
faang.groupby(['ticker', pd.Grouper(freq='1M')]).agg({
    'open': 'mean',
    'high': 'max', 
    'low': 'min', 
    'close': 'mean', 
    'volume': 'sum'
}).head()
```

```python
covid.sort_index(inplace=True)
```

```python
covid[covid.countriesAndTerritories.isin(top_five_countries)]\
.groupby(['countriesAndTerritories', pd.Grouper(freq='D')]).cases.sum().unstack(0).diff().rolling(7).mean().last('7D')
```

```python
covid.reset_index()\
    .pivot(index='date', columns='countriesAndTerritories', values='cases')\
    .drop(columns='China')\
    .fillna(0)\
    .apply(lambda x: x[(x > 0)].idxmin())\
    .sort_values()\
    .rename(lambda x: x.replace('_', ' '))
```

- e) Rank the countries by cumulative cases using percentiles.

```python
pct_cases = covid.groupby('countriesAndTerritories')['cases'].apply(sum)\
    .rename('cases_pct').transform('rank', method='max', pct=True).reset_index()\
    .sort_values('cases_pct', ascending=False)\
    .set_index('countriesAndTerritories')
```

```python
pct_cases.head()
```

```python
pct_cases['cases_pct'] = pct_cases['cases_pct'].apply(lambda x: format(x, '.2%'))
```
