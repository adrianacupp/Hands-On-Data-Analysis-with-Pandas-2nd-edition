3. Using the faang.csv file, group by the ticker and resample to monthly frequency.
Make the following aggregations:
- a) Mean of the opening price
- b) Maximum of the high price
- c) Minimum of the low price
- d) Mean of the closing price
- e) Sum of the volume traded

```python
faang = pd.read_csv('exercises/faang.csv', index_col='date', parse_dates=True)
faang.head()
```

```python
faang.groupby(['ticker', pd.Grouper(freq='1M')]).agg({
    'open': 'mean',
    'high': 'max', 
    'low': 'min', 
    'close': 'mean', 
    'volume': 'sum'
}).head()
```

4. Build a crosstab with the earthquake data between the tsunami column and the magType column. Rather than showing the frequency count, show the maximum magnitude that was observed for each combination. Put the magnitude type along the columns.

```python
df.head()
```

```python
pd.crosstab(
    index=df.tsunami,
    columns=df.magType,
    colnames=['magnitude'],
    values=df.mag,
    aggfunc=np.max 
)
```

8. Add event descriptions:
- a) Create a dataframe with the following three columns: ticker, date, and event. The columns should have the following values:
i) ticker: 'FB'
ii) date: ['2018-07-25', '2018-03-19', '2018-03-20']
iii) event: ['Disappointing user growth announced after close.', 'Cambridge Analytica story',
'FTC investigation']
- b) Set the index to ['date', 'ticker'].
- c) Merge this data with the FAANG data using an outer join.

```python
fb = pd.DataFrame(index=['2018-07-25', '2018-03-19', '2018-03-20'], columns= ['ticker', 'event'])
fb.head()
```

```python
fb = {'ticker': ['FB'], 'date': pd.to_datetime(
         ['2018-07-25', '2018-03-19', '2018-03-20']),
      'event': ['Disappointing user growth announced after close.', 'Cambridge Analytica story', 'FTC investigation']
      }
fb = pd.DataFrame(data=fb, index=[0,1,2])
fb
```

```python
fb.set_index(['date', 'ticker'], inplace= True)
fb
```

```python
new = faang.reset_index().set_index(['date', 'ticker']).join(
    fb, how='outer'
)
new.head()
```

```python
new.loc[new['event'].notnull()]
```

```python
# alternatively with query
new.query('event == event')
```

```python
# filtering not null values
new[new.event.notna()]
```
